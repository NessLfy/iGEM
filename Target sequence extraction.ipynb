```python
import pandas as pd 
from Bio import ExPASy,SeqIO
import io
```

```python
df = pd.read_excel('Target sequences oyster.xlsx', sheet_name='Feuil2', header=0)
df.head()
```

```python
from Bio import ExPASy
from Bio import SwissProt


seq_list = []
EMBL_id = []
for ID in df['Accession ID']:
    
    handle = ExPASy.get_sprot_raw(ID)
    record = SwissProt.read(handle)
    seq_list.append(record.sequence)
    EMBL_id.append(record.gene_name[40:50])
    
df['Sequence'] = seq_list
df['EMBL_id'] = EMBL_id
df.head()
```

```python
from Bio import Entrez
from urllib.error import URLError

Entrez.email = "ness.louaf@gmail.com"
list_seq = []
for i in df['Name']:
    try:
        with Entrez.esearch(db="nucleotide", retmax=1, term="Crassostrea gigas[ORGN]"+ str(i), idtype="acc")as handle:
            try:
                seq = Entrez.efetch(db="nucleotide", rettype="fasta", retmode="text", id=str(Entrez.read(handle)['IdList'][0]))
                seq_record = SeqIO.read(seq, "fasta")
                list_seq.append(seq_record.seq)
                print(str(i)+ ' is done !')

            except:
                list_seq.append('NO sequences found')
                print(str(i)+' didnt work')
    except URLError:
        print(i)
        

        
```

```python
df['DNA_seq'] = list_seq
df
```

```python
df.to_csv('Target_sequences_full.csv', index = False, encoding='utf-8')
df.to_excel('Target_sequences_full.xlsx', index = False, encoding='utf-8')
```
